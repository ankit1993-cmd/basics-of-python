 6. why are lists are mutables,but tuples are immutable ?

 9. what is type casting in python ?

    int(): converts to integers.

 10. what is difference between implicit and explicit type casting ?
    - in implicit type casting, the programming language automatically converts data types to another if needed. for example if you have an integer variable and you try to assign it to a f;loat variable, the programming language automatically converts the inetgers to a float without you having to do anythiong.
    other hand explicit type casting, also known as type coeration,occurs when the programmer explicitly converts a value from one data type to another
    unlike implicit type casting, explicit type casting requires the programmer to specify the desierd data type conversion.

 11. what is the purpose of conditional statement in python ?
    - in python, conditional statement are used to execute certain blocks of code based on specific condition, helping to control the flow of a program and make it behave differently in different situation
    these statement allow the program to react differently depending on whether a condition is True or False
    common condition statement in python include "if" "elif" and "else", whioch enable the selection of specific blocks of code to execute based on condition.

 12. how does the elif statement works in python ?
    - in python, the elif statement, which stands for "else if", is used to check multiple condition sequentially. if the first condition is false python moves on the next elif statement to check if that condition is true . if none of the conditions are true, the code will execute the else statement if it exist.
    the structure of an "if elif else" statement as follow-
    if condition 1:
        execute code if condition 1 is true
    elif condition2:
        execute code if condition 2 is true
    else:
        execute code if all conditions are false

  13. what is difference between for and while loop ?
      - for loop iterates over sequence(list,strings etc) or fixed iteration. use when you know the number of loops upfront
      - syntax-
       for i in range(3):
         print(i)

      - while loop runs while a condition is true . use for unknown iteration (e.g waiting for user input)
      - syntax-
      count=0
      while count < 3:
         print(count)
         count+= 1
