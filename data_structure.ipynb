13. how do sets handle duplicate values in python ?

    - in python,sets automatically handle duplicate values by ensuring each element is unique.here's how they manage duplicates
    1. uniqueness enforcement- when you attempt to add duplicate values to set ignores the duplicates.only one instance of each unique element is retained.
    2. sets use hash tables for storage .when adding an element,python first compute its hash (via **hash**) to locate a storage bucket.
    3. if multiple elements share the same hash(collision),python checks equality (via **eq**) to confirm if they are truely duplicates.
    4. sets are unordered,so the original insertion order is not preserved.
    5. element must be immutable and hashable.mutable can not be added to sets.
